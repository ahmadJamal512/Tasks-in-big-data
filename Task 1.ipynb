```python
from pyspark import SparkConf,SparkContext
from datetime import datetime
```

```python
conf=SparkConf().setMaster('local').setAppName('')
sc=SparkContext(conf=conf)
```

```python
customersRdd=sc.textFile('/content/customers.csv')
```

```python
customersListRdd=customersRdd.map(lambda x:x.split(','))
```

```python
customersListRdd.take(5)
```

```python
customersPairRdd=customersListRdd.map(lambda x:(x[0],x[1]))
```

```python
customersPairRdd.take(5)
```

```python
productsRdd=sc.textFile('/content/products.csv')
```

```python
productsListRdd=productsRdd.map(lambda x:x.split(','))
```

```python
productsListRdd.take(5)
```

```python
productsPairRdd=productsListRdd.map(lambda x:(x[0],x[1]))
```

```python
productsPairRdd.take(5)
```

```python
salesRdd=sc.textFile('/content/sales.csv')
```

```python
salesListRdd=salesRdd.map(lambda x:x.split(','))
```

```python
salesListRdd.take(5)
```

```python
salesPairRdd=salesListRdd.map(lambda x:(x[1],x[2]))
```

```python
salesPairRdd.take(5)
```

```python
distinctRdd = salesPairRdd.distinct()
```

```python
distinctRdd.take(5)
```

```python
countRdd=distinctRdd.map(lambda x: (x[0],1))
```

```python
countRdd.take(5)
```

```python
countTotalRdd=countRdd.reduceByKey(lambda x,y:x+y)
```

```python
countTotalRdd.take(5)
```

```python
filterRdd=countTotalRdd.filter(lambda x:x[1]>10)
```

```python
filterRdd.take(5)
```

```python
finalRdd=filterRdd.join(customersPairRdd).map(lambda x:x[1]).sortByKey(ascending=True)
```

```python
finalRdd.saveAsTextFile('/content/customers_more_than_10_products_sorted')
```
